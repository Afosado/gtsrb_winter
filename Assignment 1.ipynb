```python
import cv2
from PIL import Image
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import torch
import torch.nn.functional as F
```

```python
def image_arrs(paths, im_arr):
    for path in paths:
        #read images, may need to change filepath
        image = cv2.imread('gtsrb-german-traffic-sign/'+path)
        image_from_array = Image.fromarray(image, 'RGB')
        #resize all images to 30x30
        size_image = image_from_array.resize((30, 30))
        #convert images to array
        image_arr = np.array(size_image.getdata(),
                        np.uint8).reshape(size_image.size[1], size_image.size[0], 3)
        im_arr.append(image_arr.flatten())
    return im_arr
```

```python
df = pd.read_csv('gtsrb-german-traffic-sign/Train.csv')
```

```python
paths1 = df[(df['ClassId']==1)]['Path'].values
paths2 = df[(df['ClassId']==2)]['Path'].values
```

```python
class1_arrs = []
class1_arrs = image_arrs(paths1, class1_arrs)

class2_arrs = []
class2_arrs = image_arrs(paths2, class2_arrs)
```

```python
X = class1_arrs + class2_arrs
y = len(class1_arrs)*[0] + len(class2_arrs)*[1]
```

```python
#split data
X_train, X_test, y_train, y_test = train_test_split(
    np.array(X), np.array(y), test_size=0.2, random_state=0)
```

```python
X_train = torch.from_numpy(X_train).float()
y_train = torch.from_numpy(y_train).float().view(-1, 1)
X_test = torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test).float().view(-1, 1)
```

```python
class LR_model(torch.nn.Module):
    def __init__(self):
        super(LR_model,self).__init__()
        self.linear = torch.nn.Linear(2700,1)
    def forward(self,x):
        y_pred = F.sigmoid(self.linear(x))
        return y_pred
```

```python
lr_model = LR_model()
criterion = torch.nn.BCELoss(size_average = True)
optimizer = torch.optim.Adam(lr_model.parameters(),lr = 0.0001)
```

```python
for epoch in range(1000):
    y_pred = lr_model(X_train)
    
    loss = criterion(y_pred, y_train)
    print(epoch, loss)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
